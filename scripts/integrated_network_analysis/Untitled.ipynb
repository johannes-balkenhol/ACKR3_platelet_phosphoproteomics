```python
import pandas as pd

# Try ISO-8859-1 first (most common for German bank CSVs)
df = pd.read_csv("CHK_957.csv", sep=";", engine="python", header=None, encoding="ISO-8859-1")

# Define keywords to search for
keywords = ["miete", "rent", "steimel"]

# Combine all columns into a single searchable string per row
df["combined"] = df.astype(str).agg(" ".join, axis=1)

# Filter rows that contain any keyword (case-insensitive)
mask = df["combined"].str.contains("|".join(keywords), case=False, na=False)
filtered = df[mask]

# Save the filtered results to a new CSV file
filtered.to_csv("filtered_miete_rent_steimel.csv", sep=";", index=False, header=False, encoding="utf-8")

# Optional: print results to console
print(f"✅ Found {filtered.shape[0]} matching rows.")
print(filtered.head())
```
